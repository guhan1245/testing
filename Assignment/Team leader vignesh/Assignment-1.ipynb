## Python Basics

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/House Price India.csv")
```

```python
df.isnull().sum()
```

```python
df['number of bedrooms'].fillna(df['number of bedrooms'].median(), inplace=True)
```

```python
df.describe()
```

```python
plt.hist(df['Price'])
plt.show()
```

```python
sns.scatterplot(x='living area', y='Price', data=df)
plt.show()
```

```python
sns.boxplot(x='waterfront present', y='Price', data=df)
plt.show()
```

```python
sns.pairplot(data=df, vars=['Price', 'living area', 'number of bathrooms'])
plt.show()
```
